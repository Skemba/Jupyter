```python
import pandas as pd
```

```python
stats = pd.read_csv('datasets/NBA/Seasons_Stats.csv', index_col=0)
```

```python
stats.head()
```

```python
stats.info()
```

```python
stats.drop(stats[stats.Year<1990].index, inplace=True)
stats.drop(stats[stats.G<50].index, inplace=True)
stats.drop(stats[stats.MP<1000].index, inplace=True)
stats.dropna(subset=['Player'], inplace=True)
stats.dropna(subset=['PER'], inplace=True)
stats.dropna(subset=['Tm'], inplace=True)
stats.dropna(subset=['Pos'], inplace=True)

stats = stats[[ 'Player', 'PER', 'TS%', 'OWS', 'WS', 'DWS', 'TRB%', 'USG%', 'PTS', 'G']]
stats
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
stats.hist(column=['PER', 'TS%', 'OWS', 'WS', 'DWS', 'TRB%', 'USG%'], bins=50, figsize=(20,15))
plt.show()
```

```python
import numpy as np

stats["PER_cat"] = np.ceil(stats["PER"]/ 5)
stats.PER_cat.value_counts(dropna=False)
```

```python
stats.hist(column=['PER_cat'], bins=50, figsize=(20,15))
plt.show()
```

```python
stats[stats.Player.str.startswith('Nikola J')].head()
```

```python
stats['PPG'] = stats.PTS / stats.G
stats[(stats.PER > 26) & (stats.PPG < 25)].head(30)
```

```python
stats[stats.PER > 26].Player.value_counts()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(stats, stats['PER_cat']):
    print (train_index.max())
    strat_train_set = stats.iloc[train_index]
    strat_test_set = stats.iloc[test_index]
```

```python
for set in (strat_train_set, strat_test_set):
    set.drop(["PER_cat"], axis=1, inplace=True)
```

```python
stats = strat_train_set.copy()
```

```python
stats.plot(kind="scatter", x="PER", y="PPG", alpha=0.1)
```

```python
stats.describe()
```

```python
corr_matrix = stats.corr()
corr_matrix.PPG.sort_values(ascending=False)
```

```python
stats = strat_train_set.drop("PPG", axis=1)
stats_labels = strat_train_set["PPG"].copy()
```

```python
from sklearn.preprocessing import Imputer

imputer = Imputer(strategy="median")

stats_num = stats.drop(['Player', 'PTS', 'G', 'USG%', 'DWS', 'OWS'], axis = 1)

imputer.fit(stats_num)
imputer.statistics_
# imputer and other estimators only "estimate" (do mathematical operations, for example find median) some values based on data fed to them with method fit()
# transformers are estimators that can transform data given the estimations they derived with method transform()
# predictors are estimators capable of making predictions of some new data instance with method predict()
```

```python
# now use "trained" imputer to transform training set
X = imputer.transform(stats_num)
```

```python
# result of imputer is numpy array, so if we want we can transform it to dataframe 
stats_tr = pd.DataFrame(X, columns=stats_num.columns)
```

```python
stats_tr.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', Imputer(strategy="median")),
    ('std_scaler', StandardScaler()),
])

stats_num_tr = num_pipeline.fit_transform(stats_num)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(stats_num_tr, stats_labels)
```

```python
some_data = stats_num.iloc[:5]
some_labels = stats_labels.iloc[:5]
some_data_prepared = num_pipeline.transform(some_data)

print("Predictions:", lin_reg.predict(some_data_prepared))
```

```python
print("Labels:", list(some_labels))
```

```python
dd = stats_num[stats_num.PER == 26.4]
lbl = stats_labels.iloc[[stats_num.index.get_loc(dd.iloc[0].name), stats_num.index.get_loc(dd.iloc[1].name), stats_num.index.get_loc(dd.iloc[2].name)]]

print(lin_reg.predict(num_pipeline.transform(dd)))

print(lbl)
```
