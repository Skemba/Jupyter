```python
# Objective of this project is to see how successful can machine learning algorithms be in detecting intrusion detection.
# Supervised learning makes room for error much smaller than unsupervised learning.
# Most cyber security problems should be approached with great caution, thus supervised learning is preferred.
# Batch learning will be used as training method because there is no constant flow of data needed for online learning.
# Also, online learning can be very harmful for security based systems if corrupt data starts being fed to the algorithm.
# This is a classification problem, it is meant to group different types of network traffic.
# Most importantly it should classify between malicious and non-malicious. 
# As secondary task it should classify exactly which class of malicious attacks traffic belongs to.
# DARPA's KDD CUP 99 dataset will be used because there are no recent publicly available datasets that are large enough.
```

```python
from sklearn.datasets import fetch_kddcup99
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

```python
# Get the data
kddcup = fetch_kddcup99(subset='SA', percent10=True)
kddcup
```

```python
X, y = kddcup["data"], kddcup["target"]
```

```python
# Modify data so it would be pandas compatible
Xframe = pd.DataFrame(X)
yframe = pd.DataFrame(y)
Xframe.columns = ["duration","protocol_type","service","flag","src_bytes",
    "dst_bytes","land","wrong_fragment","urgent","hot","num_failed_logins",
    "logged_in","num_compromised","root_shell","su_attempted","num_root",
    "num_file_creations","num_shells","num_access_files","num_outbound_cmds",
    "is_host_login","is_guest_login","count","srv_count","serror_rate",
    "srv_serror_rate","rerror_rate","srv_rerror_rate","same_srv_rate",
    "diff_srv_rate","srv_diff_host_rate","dst_host_count","dst_host_srv_count",
    "dst_host_same_srv_rate","dst_host_diff_srv_rate","dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
    "dst_host_rerror_rate","dst_host_srv_rerror_rate"]
```

```python
# Next few lines provide a quick look at the data structure.
```

```python
Xframe.head()
```

```python
Xframe.describe()
```

```python
Xframe.info()
```

```python
# Before gaining more insight into the data, split train and test data to avoid possible later bias.
from sklearn.model_selection import train_test_split

train_x, test_x = train_test_split(Xframe, test_size=0.2, random_state=42)
train_y, test_y = train_test_split(yframe, test_size=0.2, random_state=42)
```

```python
# Since all attributes are objects (strings) turn them into numeric if that's their underlying type.
train_x = train_x.apply(pd.to_numeric, errors='ignore')
```

```python
# Check attribute types after applying to_numeric.
train_x.info()
```

```python
# Separate numeric data from categorical data for now.
protocol_type = train_x.protocol_type
service = train_x.service
flag = train_x.flag

train_x = train_x.drop(['protocol_type', 'service', 'flag'], axis=1)
```

```python
# We will now try to visualize data, mostly to remove unnecessary attributes and attributes with high correlation.
```

```python
colormap = plt.cm.viridis
plt.figure(figsize=(15,15))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(train_x.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
# Correlations:
# (dst_host_serror_rate, serror_rate, srv_serror_rate, dst_host_srv_serror_rate)
# (rerror_rate, srv_rerror_rate, dst_host_rerror_rate, dst_host_srv_rerror_rate)
# (count, srv_count)
# (num_root, num_compromised)
# (dest_host_srv_count, dst_host_same_srv_rate)
# (is_guest_login, hot)

# We will drop all but one from each group, leaving the most general attribute.

train_x = train_x.drop(['dst_host_srv_serror_rate', 'srv_serror_rate', 'dst_host_serror_rate', 'srv_count', 'num_compromised', 'srv_rerror_rate', 'dst_host_rerror_rate', 'dst_host_srv_rerror_rate', 'dst_host_srv_count', 'is_guest_login'], axis=1)
```

```python
train_x.is_host_login.value_counts()
```

```python
train_x.num_outbound_cmds.value_counts()
```

```python
# From previous two commands we can see that both is_host_login and num_outbound_cmds have value of 0 in all cases.
# This allows us to remove them both without losing performance.
train_x = train_x.drop(['is_host_login', 'num_outbound_cmds'], axis=1)
```

```python
colormap = plt.cm.viridis
plt.figure(figsize=(15,15))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(train_x.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
# Correlation matrix graph is much clearer now and there are no more strong correlations that we need to remove.
```

```python
# Next step would be handling the categorical data. We will use LabelEncoder and OneHotEncoder for this.
# Label encoder will assign integer number to each class inside of categorical data.
# OneHotEncoder will then make two-dimensional array that is more appropriate for algorithm input.
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
protocol_1hot = encoder.fit_transform(protocol_type)
protocol_1hot
```

```python
print(encoder.classes_)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder() 
protocol_1hot = encoder.fit_transform(protocol_1hot.reshape(-1,1))
protocol_1hot.toarray()
```

```python
encoder = LabelEncoder()
service_1hot = encoder.fit_transform(service)
service_1hot
```

```python
encoder = OneHotEncoder()
service_1hot = encoder.fit_transform(service_1hot.reshape(-1,1))
service_1hot.toarray()
```

```python
service_1hot.toarray().shape
```

```python
encoder = LabelEncoder()
flag_1hot = encoder.fit_transform(flag)
flag_1hot
```

```python
encoder = OneHotEncoder()
flag_1hot = encoder.fit_transform(flag_1hot.reshape(-1,1))
flag_1hot.toarray()
```

```python
# Since label is also categorical we will turn it into integer as well for easier comparing
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_1hot = encoder.fit_transform(train_y.values.ravel())
y_1hot
```

```python
# Not all attributes will be using the same scale, some won't even be in the same order of magnitude.
# This can cause some attributes to be given higher significance level based on their scale.
# We will use standardization feature scaling as preprocessing before feeding data to machine learning algorithm.
# Other common method is min-max scaling which gives more significance to outliers.
# We don't want outliers to be more significant because it could make us miss some subtle differences between malicious and non-malicious network traffic.
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_x_scaled = scaler.fit_transform(train_x, y_1hot)
```

```python
# Turn data into shape suitable for algorithm input (pandas dataframe isn't valid input)
train_x_prepared = np.c_[train_x_scaled, protocol_1hot.toarray(), service_1hot.toarray(), flag_1hot.toarray()]
```

```python
train_x_prepared.shape
```

```python
# Try a few machine learning algorithms and see which one of them performs best based on accuracy of predicted results.
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.cross_validation import cross_val_score

classifier = GaussianNB()
cross_val_score(classifier, train_x_prepared, y_1hot, cv=3, scoring='accuracy')
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(max_depth=5)
cross_val_score(classifier, train_x_prepared, y_1hot, cv=3, scoring='accuracy')
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1)
cross_val_score(classifier, train_x_prepared, y_1hot, cv=3, scoring='accuracy')
```

```python
from sklearn.svm import SVC

classifier = SVC(kernel="linear", C=0.025)
cross_val_score(classifier, train_x_prepared, y_1hot, cv=3, scoring='accuracy')
```

```python
from sklearn.neural_network import MLPClassifier

classifier = MLPClassifier(alpha=1)
cross_val_score(classifier, train_x_prepared, y_1hot, cv=3, scoring='accuracy')
```

```python
# SVC, DecisionTree and MLP all have accuracy of ~99.9% which seems like a really good result.
# In this case, though, what's more important than accuracy is precision.
# It's less costly for security systems to classify non-malicious traffic as malicious than malicious as non-malicious
# So we would want the number of "false positives" to be as low as possible

# To do this we will turn all the results to only classify between malicious and non-malicious rather than give each malicious traffic it's specific class.
```

```python
# Let's see classes that LabelEncoder has give to results
print(encoder.classes_)
```

```python
# Normal traffic is at index 5, so we will set everything with index 5 as True, otherwise False.
y_1hot_bool = (y_1hot == 5)
```

```python
# Now we find precision (true_positive/true_positive+false_positive) for three classifiers that had accuracy of ~99.9%
```

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.model_selection import cross_val_predict
classifier = DecisionTreeClassifier(max_depth=5)
train_y_pred = cross_val_predict(classifier, train_x_prepared, y_1hot_bool, cv=3)
precision_score(y_1hot_bool, train_y_pred)
```

```python
classifier = SVC(kernel="linear", C=0.025)
train_y_pred = cross_val_predict(classifier, train_x_prepared, y_1hot_bool, cv=3)
precision_score(y_1hot_bool, train_y_pred)
```

```python
classifier = MLPClassifier(alpha=1)
train_y_pred = cross_val_predict(classifier, train_x_prepared, y_1hot_bool, cv=3)
precision_score(y_1hot_bool, train_y_pred)
```

```python
# What we get as a result for precision seems to be great, but in security even 99.9999% security still isn't 100% secure.
# That's why it's important to keep working on data cleaning, preprocessing and algorithms themselves.

# What can be concluded is that machine learning can be pretty effecient at intrusion detection.
# What's also visible is that neural nets approach (MLP) does not give the best results in this case.
# It could be because of bad hyperparameter choice (GridSearchCV can be used to find the optimal ones, but it requires lots of resources).
# But it could also be that traditional machine learning algorithms offer better performance in cyber security, which is actually noticed in a few scientific papers.
# One suggestion I would add is that MLP might've worked better if before training it we've done some rule-based removal of outliers or if we gave output of more traditional algorithm as input of MLP.

# What was left to be done is for model to be tested on test data, but that would require doing all the same transformations as on training data, so it won't be done here.
# Results might be a little less satisfactory on test data, but it wouldn't be enough to disprove everything stated before.
```
